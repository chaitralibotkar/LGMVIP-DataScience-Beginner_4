#### Importing necessary libraries

```python
#Importing the opencv and matplotlib library
import cv2
import matplotlib.pyplot as plt
```

#### Loading image 

```python
#Locating the image
image_location='E:/Internship/Letgrowmore/Pencil Sketch/img4.jpg'
```

#### Reading the image with openCV library 

```python
#Reading the image
image=cv2.imread(image_location)
```

#### Showing the image 

```python
#Showing the image
plot_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(plot_image)
plt.axis(False)
plt.show()
```

#### Converting original image to grayscale image 

```python
#Converting the original image to grayscale image
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
RGB_image=cv2.cvtColor(gray_img,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.axis(False)
plt.show()
```

#### Converting image to negative

```python
#Converting image to negative
inverted_gray_image=255-gray_img
RGB_image=cv2.cvtColor(inverted_gray_image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.axis(False)
plt.show()
```

#### Blurring image to smooth the process 

```python
#Blurring the image for smooth image
blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21),0) 
RGB_image=cv2.cvtColor(blurred_img,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.axis(False)
plt.show()
```

#### Producing negative of blurred image 

```python
#Inverting i.e. producing negative of the blurred image
inverted_blurred_img = 255 - blurred_img
RGB_image=cv2.cvtColor(inverted_blurred_img,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.axis(False)
plt.show()
```

#### Converting image to pencil sketch 

```python
#Converting image to pencil sketch
pencil_sketch = cv2.divide(gray_img, inverted_blurred_img, scale = 256.0)
pencil_sketch_img=cv2.cvtColor(pencil_sketch,cv2.COLOR_BGR2RGB)
plt.imshow(pencil_sketch_img)
plt.axis(False)
plt.show()
```

#### Comparision of original image and pencil sketch 

```python
#Plotting both images side by side for comparision
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
plt.title("Original photo")
plt.imshow(plot_image)
plt.axis(False)

plt.subplot(1,2,2)
plt.title("Pencil Sketch")
plt.imshow(pencil_sketch_img)

plt.axis(False)
plt.show()
```
